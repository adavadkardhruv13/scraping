```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.support import expected_conditions as EC

service = Service('C:\Program Files (x86)\chromedriver-win32\chromedriver.exe')
driver = webdriver.Chrome(service=service)
driver.implicitly_wait(1)

product_name = []
product_price = []
product_ratings = []
product_ratings_num = []
product_link = []

for page in range(1, 21):
    driver.get(f'https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_{page}')
    items = wait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//div[contains(@class, "s-result-item s-asin")]')))
    for item in items:
        # For name of item
        name = item.find_element(By.XPATH, './/span[@class="a-size-medium a-color-base a-text-normal"]')
        product_name.append(name.text)

        # For price
        whole_price = item.find_elements(By.XPATH, './/span[@class="a-price-whole"]')
        if whole_price:
            price = '.'.join([whole_price[0].text])
        else:
            price = 0
        product_price.append(price)

        # For ratings box
        ratings_box = item.find_elements(By.XPATH, './/div[@class="a-row a-size-small"]/span')

        # For ratings and ratings_num
        if ratings_box:
            ratings = ratings_box[0].get_attribute('aria-label')
            ratings_num = ratings_box[1].get_attribute('aria-label')
        else:
            ratings, ratings_num = 0, 0

        product_ratings.append(ratings)
        product_ratings_num.append(str(ratings_num))

        # For product link
        link = item.find_element(By.XPATH, './/a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]').get_attribute("href")
        product_link.append(link)

    if page < 20:
        next_button = item.find_element(By.XPATH, '//*[@id="search"]/div[1]/div[1]/div/span[1]/div[1]/div[31]/div/div/span/a[3]')
        next_button.click()
        time.sleep(5) 

# Print the data after scraping all 20 pages
#print(product_name)
#print(product_price)
#print(product_ratings)
#print(product_ratings_num)
#print(product_link)

description = []
asin = []
product_description = []
manufacturer = []

for link in product_link:
    driver.get(link)
    time.sleep(2)  
    
    #part2
    
    description_value = item.find_element(By.XPATH, '//*[@id="feature-bullets"]/ul/li[1]/span')
    description_value.append("description_value")
      
    asin_value = item.find_element(By.XPATH, '//*[@id="detailBullets_feature_div"]/ul/li[4]/span/span[2]')                                
    asin_value.append("asin_value")
                                   
    product_description_value = item.find_element(By.XPATH,'//*[@id="productFactsDesktopExpander"]/div[1]/ul/span/li/span')                               
    product_description_value.append("product_description_value")
                         
    manufacturer_value = item.find_element(By.XPATH, '//*[@id="detailBullets_feature_div"]/ul/li[8]/span/span[2]/text()')                    
    manufacturer_value.append("manufacturer_value")

# Step 2: Combine the data into a DataFrame
product_data = {
    "Product Name": product_name,
    "Product Price": product_price,
    "Product Ratings": product_ratings,
    "Product Ratings Count": product_ratings_num,
    "Product Link": product_link,
    "Description": description,
    "ASIN": asin,
    "Product Description": product_description,
    "Manufacturer": manufacturer
}
```

```python
df = pd.DataFrame(product_data)

# Step 3: Export the data to a CSV file
df.to_csv("data.csv", index=False)
```
